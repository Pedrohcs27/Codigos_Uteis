```python
# Importção das bibliotecas
import pandas as pd
import geopandas as gpd
import difflib as dif
```

```python
# Leo o caminho do shape>.

x = gpd.read_file('dados/dado.shp')
```

```python
# Cria dataframe com as colunas de interesse.
y = x[['MUNICIPIO','MUNICN']]
y
```

```python
# Definição de função para saber porcentagem de semelhanças entre duas colunas de texto.
def sml(x,y):
    return SequenceMatcher(None, x, y).ratio()
```

```python
#Cria lista com porcentagem de semelhanças entre as duas colunas.
from difflib import SequenceMatcher
lista = []
for i in range(len(x)):
    um = x.MUNICIPIO[i]
    dois = x.MUNICN[i]
    lista.append(sml(um,dois))
lista
```

```python
# Contagem de semelhanças com porcentagem maior que 87,5%.
# Porcentagem baseada na análise das principais diferenças observadas na "lista".
correto = 0
errado = 0

for j in lista:   
    correto = correto + 1 if j >= 0.875 else correto
    errado = errado + 1 if j < 0.875 else errado

print('Numero de municipios corretos é igual a',correto)
print('Numero de municipios errados é igual a',errado)


            
```

```python
# Cria um dataframe com as porcentagens de semelhança a partir da 'lista'.
z = pd.DataFrame(lista,columns = ['Comparac'])
```

```python
# Faz a união da coluna de porcentagem de com o dataframe de colunas de interesse.
m = pd.merge(y,z, how = 'inner',left_index=True, right_index=True)
```

```python
m
```

```python
# Agrupa os dados pela coluna de porcentagem de semalhança com a contagem de ocorrências.
m.groupby(['Comparac']).count()
```
